```python
import pandas as pd
```

```python
sport = pd.read_csv("season_goals.csv")
```

```python
sport
```

```python
for i in sport:
    print(f"{i}: {sport[i].isna().mean()}")
```

```python
sport = sport.dropna()
```

```python
for i in sport:
    print(f"{i}: {sport[i].isna().mean()}")
```

```python
sport.count()
```

```python
sport['position'] = sport['position'].replace({
    "C": "Centre",
    "LW": "Leftwing",
    "RW": "Rightwing",
    "D": "Defence"
})
```

```python
def calculate_total_years(year_range):
    start_year, end_year = map(int, year_range.split('-'))
    if start_year < 2000:
        end_year += 1900
    else:
        end_year += 2000
    
    return end_year - start_year + 1
```

```python
sport['years_active'] = sport['years'].apply(calculate_total_years)
```

```python
sport.columns
```

```python
column_order = ['rank', 'position', 'hand', 'player', 'years', 'years_active', 'total_goals', 'status',
       'yr_start', 'season', 'age', 'team', 'league', 'season_games', 'goals',
       'assists', 'points', 'plus_minus', 'penalty_min', 'goals_even',
       'goals_power_play', 'goals_short_handed', 'goals_game_winner',
       'headshot']
```

```python
sport = sport[column_order]
```

```python
sport
```

```python
sport.to_csv("new_season_goals.csv", index= False)
```
